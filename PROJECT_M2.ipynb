```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# loading Starbucks datasets:
starbuck_drink_menu = pd.read_csv('starbucks_drinkMenu_expanded.csv', encoding='utf-8')
starbuck_drink_nutri = pd.read_csv('starbucks-menu-nutrition-drinks.csv',  encoding='utf-8')
starbuck_food_nutri = pd.read_csv('starbucks-menu-nutrition-food.csv', encoding='utf-16')

#loading Macdonald's dataset
macdonald_menu = pd.read_csv('menu.csv',  encoding='utf-8')

# Adding missing column name as Food_Name:
starbuck_food_nutri.rename(columns ={'Unnamed: 0': 'Food_Name'}, inplace =True)

# Adding missing column name as Drink_Name:
starbuck_drink_nutri.rename(columns ={'Unnamed: 0': 'Drink_Name'}, inplace =True)
```

#### BUSINESS CASE 1: Most asked "Caffeine", "Sodium", and "Sugar" nutritional facts in Starbucks Drink Items 

```python
#plot visualization for caffeine contents: 

fig, ax = plt.subplots()
ax.grid(True)
plt.style.use('ggplot')
x = starbuck_drink_menu["Beverage_prep"] [:4]
y = starbuck_drink_menu["Caffeine (mg)"] [:4]
plt.title("Caffeine (mg) in Starbucks Drink Items")
plt.xlabel("Beverage")
plt.ylabel("Caffeine (mg)")
plt.plot(x, y, marker="o")
plt.rcParams['figure.figsize'] = (10,6)
plt.show()
fig.tight_layout()
```

```python
fig, ax = plt.subplots()
plt.style.use('seaborn-poster')
plt.rcParams['figure.figsize'] = (10,6)
plt.hist(starbuck_drink_menu[" Sodium (mg)"])
plt.title("Sodium(mg) in Starbucks Drint Items")
plt.xlabel('Sodium(mg) distribution')
plt.ylabel('Counts')
fig.tight_layout()
```

```python
fig, ax = plt.subplots()
plt.style.use('seaborn-poster')
plt.rcParams['figure.figsize'] = (10,6)
plt.hist(starbuck_drink_menu[" Sugars (g)"])
plt.title("Sugars(gm) in Starbucks Drink Items")
plt.title("Sugars(gm) in Starbucks Menu Items")
plt.xlabel('Sugars(gm) distribution')
plt.ylabel('Counts')
fig.tight_layout()
```

#### BUSINESS CASE 2: The nutrition of food item(max calories, max fat(g) and max carbs(g) in Starbucks Food Items

```python
#function to return nutrition contents: 
def food_nutri_details(dataframe, food_name):
    """This function returns nutrition contents.

       parameter : dataframe : panda dataframe object
   
       parameter : food_name : name of item
   
       return : list of nutrtion contents.

    """
    for index, row in dataframe.iterrows():
        if row[0] == food_name:
            return row
```

```python
#visulization for Food nutrition items:

food_calories = [ row for i, row in starbuck_food_nutri.iterrows() if row[1] == starbuck_food_nutri[' Calories'].max()] 
food_fat = [ row for i, row in starbuck_food_nutri.iterrows() if row[2] ==  starbuck_food_nutri[' Fat (g)'].max()]
food_carbs = [ row for i, row in starbuck_food_nutri.iterrows() if row[3] == starbuck_food_nutri[' Carb. (g)'].max()]

nutrients = starbuck_food_nutri.columns[1:]

#food item with max calories 
cal_nutri = food_calories[0][1:]
plt.style.use('seaborn-poster')
plt.rcParams['figure.figsize'] = (10,6)
plt.bar(nutrients, cal_nutri, width=0.3, color=['blue', 'red', 'green', 'Fuchsia', 'Aqua'])
plt.title("Lentils & Vegetable Protein Bowl with Brown Rice")
plt.xlabel('Nutrients')
plt.ylabel('Nutrition in (gm)')
plt.show()

print('\n\n\n')
#food item with max fat
fat_nutri = food_fat[0][1:]
plt.style.use('seaborn-poster')
plt.rcParams['figure.figsize'] = (10, 6)
plt.bar(nutrients, fat_nutri, width=0.3, color=['blue', 'red', 'green', 'Fuchsia', 'Aqua'])
plt.title("Green Goddess Avocado Salad")
plt.xlabel('Nutrients')
plt.ylabel('Nutrition in (gm)')
plt.show()

print('\n\n\n')
#food item with max carbs
carbs_nutri = food_carbs[0][1:]
plt.style.use('seaborn-poster')
plt.rcParams['figure.figsize'] = (10,6)
plt.bar(nutrients, carbs_nutri, width=0.3, color=['blue', 'red', 'green', 'Fuchsia', 'Aqua'])
plt.title("Strawberries & Jam Sandwich")
plt.xlabel('Nutrients')
plt.ylabel('Nutrition in (gm)')
plt.show()
```

#### BUSINESS CASE 3:  Starbucks Compare to McDonald’s Food Items Terms of Nutritional Facts 

```python
# Compare starbucks Vs McDonalad's in 500 calories food items:

mcdonalds_menu = pd.read_csv('menu.csv', encoding='utf-8')

#Group-1 : 500 Calory items
md_cal = mcdonalds_menu['Calories'] == 500
md_bf = mcdonalds_menu['Category'] == 'Breakfast'
md_item = mcdonalds_menu[md_cal & md_bf]
md_nutri_500cal = [float(md_item['Total Fat']), int(md_item['Carbohydrates']), 
                   int(md_item['Dietary Fiber']), int(md_item['Protein'])]

sb_cal = starbuck_food_nutri[' Calories'] == 500
sb_item = starbuck_food_nutri[sb_cal]
sb_nutri_500cal = [float(sb_item[:1][' Fat (g)']), int(sb_item[:1][' Carb. (g)']), 
                   int(sb_item[:1][' Fiber (g)']), int(sb_item[:1][' Protein (g)'])]


Starbucks = sb_nutri_500cal
McDonald = md_nutri_500cal
index = ['Fat(g)', 'Carb(g)', 'Fibre(g)', 'Protein(g)']
plt.style.use('seaborn-poster')
plt.rcParams['figure.figsize'] = (10,6)
df = pd.DataFrame({'Starbucks': Starbucks, 'McDonald': McDonald}, index=index)
ax = df.plot.bar(rot=0)
ax.set_title("Starbucks Vs McDonald's in 500 calory food item")
ax.set_xlabel('Nutrients', fontsize=15)
ax.set_ylabel('Nutrition in (gm)', fontsize=15)
plt.tight_layout()
```
